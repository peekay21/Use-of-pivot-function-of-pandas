```python
import pandas as pd
```

Pivot allows you to transform or reshape data

```python
'''create the dataframe'''

df=pd.DataFrame({
    'date':['5/1/2017','5/2/2017','5/3/2017','5/1/2017','5/2/2017','5/3/2017','5/1/2017','5/2/2017','5/3/2017'],
    'city':['new york','new york','new york','mumbai','mumbai','mumbai','beijing','beijing','beijing'],
    'temperature':[65,66,68,75,78,82,80,77,79],
    'humidity':[56,58,60,80,83,85,26,30,35]
})
```

```python
df
```

```python
print('temperature')
print('==================================================')
print(df.pivot(index='date',columns='city',values='temperature'))

print('\nHumidity')
print('==================================================')
print(df.pivot(index='date',columns='city',values='humidity'))
```
